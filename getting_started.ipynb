**Basic For loop**

```python
years_experience ={
    'Luke':3,
    'Kelly' :6,
    'Ken':4,
    'Alex':3,
    'Mary':7
}
```

```python
for name in years_experience:
    experience = years_experience[name]
    print(f"{name} has {experience} years of experience.")
```

**Basic While loop**

```python
items = list(years_experience.items())
index = 0

# Basic while loop to iterate and check condition
while index < len(items):
    name, experience = items[index]
    if experience > 4:
        print(f"{name} has more than 4 years of experience.")
    index += 1
```

Class


```python
class BaseSalary:
  def __init__(self, base_salary, bonus_rate=0.1, symbol="$"):
    self.base_salary = base_salary
    self.bonus_rate = bonus_rate
    self.symbol = symbol
    self.total_salary = base_salary * (1 + bonus_rate)
    self.bonus = self.total_salary - self.base_salary

  def __repr__(self):
    return f'{self.symbol}{self.base_salary:,.0f}'

  def show_salary(self):
    return f'{self.symbol}{self.total_salary:,.0f}'

  def show_bonus(self):
    return f'{self.symbol}{self.bonus:,.0f}'
```

```python
salary=BaseSalary(1000)
print(salary)
```

```python
salary.show_bonus()
```

Median Salary vs Data Science Jobs

```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset= load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()

df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
```

```python
job_salary = df.groupby('job_title_short')['salary_year_avg'].median().sort_values()
```

```python
job_salary.plot(kind="barh")
plt.xlabel('Salary ($USD)')
plt.ylabel('')
plt.title('Median Salary by Job Title')
plt.show()
```
